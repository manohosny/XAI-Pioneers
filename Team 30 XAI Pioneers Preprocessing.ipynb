```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# fetch dataset
phishing_websites = fetch_ucirepo(id=327)

# Access the features and target separately
X = phishing_websites.data.features
y = phishing_websites.data.targets

# Create a DataFrame using the features
df = pd.DataFrame(X, columns=phishing_websites.data.feature_names) # Assuming feature names are available
# Add the target variable as a column if needed
df['target'] = y

# Now you can use df as your DataFrame
df.head()
```

```python
print("Dataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nTarget Distribution:")
print(df['target'].value_counts())
print("\nSummary Statistics:")
print(df.describe())
print("Duplicates:")
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```
